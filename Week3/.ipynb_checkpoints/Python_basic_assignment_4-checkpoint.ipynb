### Solution 1.2

```python
# Method 1
def check_long_words(lst):
        long_word = []
        j = 0
        for i in range(0,len(lst)//2): 
            if len(lst[j]) > lst[j+1]:
                long_word.append(lst[j])
            j = j+2
        return long_word
      
list_of_word_and_number = ['Thanga',5,'ineuron',3,'raj',1,'Bangalore',10]
result_of_long_words = check_long_words(list_of_word_and_number) 
print(result_of_long_words)
```

```python
# Method 2
def filter_long_words(words, n):
    return list(filter(lambda x: len(x) > n, words))


filter_long_words(['Thanag', 'Bangalore', 'test'], 5)
```

```python
# Method 2
def world_length_cal(n):
    return len(n)
input_list =['ab','cde','erty']
list(map(world_length_cal, input_list))
```

```python
# Method 3 using normal function
def world_length_cal(lst):
    world_length_list =[]
    if (type(lst)==list):
        for i in lst:
            world_length_list.append(len(i))    
    else:
        print('Check your input list')
    return world_length_list
```

```python
input_list =['ab','cde','erty']
world_length = world_length_cal(input_list)
world_length
```

```python
input_list =['Thangaraj','iNeuron','Machine Learning']
world_length = world_length_cal(input_list)
world_length
```
