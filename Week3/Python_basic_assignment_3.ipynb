# Solution for Problem 1.1:

```python
def my_add(lst):
     if all(isinstance(x, int) for x in lst):
       return (sum(lst))
     elif all(isinstance(x, str) for x in lst):
        result= ''
        for element in lst:
           result += str(element)
        return result
        
```

```python
def my_max(lst):
     return(max(lst))
```

```python
def my_reduce(function_name,my_seq):
    if (function_name == 'my_add') and (type(my_seq)==list):
        my_add_result = my_add(my_seq)
        return(my_add_result)
    elif(function_name == 'my_max') and (type(my_seq)==list):
        my_max_result = my_max(my_seq)
        return(my_max_result)
    else:
        print('Check your input function name and datatype')
     
```

```python
function_name = 'my_add'
list_value = [11,21,31,41,51]
my_final_result = my_reduce(function_name,list_value)    
my_final_result   
```

```python
function_name = 'my_max'
list_value = [1,25,3,4,5]
my_final_result = my_reduce(function_name,list_value)    
my_final_result   
```

# Solution for Problem 1.2:

```python
def my_number_split(lst):
    raw_even_data = []
    raw_odd_data = []
    for i in range(len(lst)):
        if (lst[i]%2 == 0) :
          raw_even_data.append(lst[i]) 
        else:
          raw_odd_data.append(lst[i])   
    return raw_even_data,raw_odd_data
```

```python
def my_string_num_split(lst):
    m =[]
    n =[]
    for i in lst:
        if i.isalpha() == True:
           m.append(i)
        if i.isdecimal() == True:
           n.append(i)
    return m,n
```

```python
def my_filter(function_name,my_seq):
    if (function_name == 'my_number_split') and (type(my_seq)==list) and (all(isinstance(x, int) for x in my_seq)):
        my_even_result,my_odd_result = my_number_split(my_seq)
        return( my_even_result,my_odd_result)
    elif(function_name == 'my_string_num_split') and (type(my_seq)==list)and (all(isinstance(x, str) for x in my_seq)) :
        my_int_result,my_string_result = my_string_num_split(my_seq)
        return(my_int_result,my_string_result)
    else:
        print('Check your input function name and datatype')
```

```python
function_name = 'my_number_split'
list_value = [11,21,31,41,51,20]
even_num,odd_num = my_filter(function_name,list_value)
print('The filtered odd number: ',odd_num)
print('The filtered even number: ',even_num)
```

```python
function_name = 'my_string_num_split'
list_value = ['123','345','Thanga','456','Raj','BE']
int_result,string_result = my_filter(function_name,list_value)    
print('The filtered odd number: ',int_result)
print('The filtered even number: ',string_result)
```

```python
function_name = 'wrong'
list_value = ['123','345','Thanga',123]
my_filter(function_name,list_value)    
```
