```python
import PyPDF2
import io
# creating a pdf file object
pdfFileObj = open('Assignment stats 1.pdf', 'rb')
 
# creating a pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
 
# printing number of pages in pdf file
print("Number of page:",pdfReader.numPages)
 
# creating a page object
pageObj = pdfReader.getPage(0)
 
# extracting text from page
line_text =[]
pages_text = pageObj.extractText()
for line in io.StringIO(pages_text,newline="\n"):
    line_text.append(line)
# Remove the new line
new_line_removed= list(map(str.rstrip,line_text))
# Remove null 
str_list = list(filter(None, new_line_removed))

# closing the pdf file object
pdfFileObj.close()
```

```python
string_values = list(str_list[5]+' '+str_list[6])
without_empty_strings = [string for string in string_values if string != ' ']
values = [int(i) for i in without_empty_strings]
values
```

### Mean

```python
# Method1
def my_mean(sample):
    return sum(sample) / len(sample)
result = my_mean(values)
print('Mean Value of given list :', result)
```

```python
# Method 2
import statistics
statistics.mean(values)
```

```python
# Method 3
import numpy as np
np.mean(values)
```

### Median

```python
def my_median(sample):
     n = len(sample)
     index = n // 2
     # Sample with an odd number of observations
     if n % 2:
         return sorted(sample)[index]
     # Sample with an even number of observations
     return sum(sorted(sample)[index - 1:index + 1]) / 2

my_median(values)
```

```python
statistics.median(values)
```

```python
np.median(values)
```

### Mode

```python
from collections import Counter

def my_mode(sample):
    c = Counter(sample)
    return [k for k, v in c.items() if v == c.most_common(1)[0][1]]

my_mode(values)
```

```python
statistics.mode(values)
```

```python
a = np.array(values)
counts = np.bincount(a)
print(np.argmax(counts))
```

## Standard deviation

```python
import math
# Method1
def my_std(sample):
    mean = sum(sample) / len(sample)
    deviations = [(x - mean) ** 2 for x in sample]
    variance = sum(deviations) / len(sample)
    return math.sqrt(variance)
result = my_std(values)
print('std Value of give list :', result)
```

```python
statistics.stdev(values)
```

```python
np.std(values)
```
