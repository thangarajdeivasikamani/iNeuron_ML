```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
print(sns.__version__)
sns.set_theme();
```

```python
df = pd.read_csv(r'C:\MachineLearning\iNeuron_ML\Week10\statistics_week1/iris.csv')
df.head()
```

```python
# Pair plot will provide co-relation
sns.pairplot(df)
```

```python
#if I use hue we will get each group in different color.
sns.pairplot(df,hue='species')
```

```python
print(np.mean(df['sepal_length']))
print(np.median(df['sepal_length']))
print(stats.mode(df['sepal_length']))
```

```python
# Above result the median & mode are similar so no outlier
# Now add outlier & check
ls = list(df['sepal_length'])
ls.append(100)
print(np.mean(ls))
print(np.median(ls))
# Now we can see the mean got increased.
# From here we can say median will be good for outlier
```

```python
# we can see outlier
sns.boxplot(ls)
```
