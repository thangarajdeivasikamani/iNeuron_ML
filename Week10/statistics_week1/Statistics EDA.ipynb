## PDF ( Probability Distribution Function to Estimate Percentiles)
- pdf is smoother version of histogram
- x axis is bin
- y axis is Percentiles value


### Cumulative Density Function

- cumulative percentiles value assume we have above pdf curver point A pdf is 0.5 point B pdf is 1.0 & C is 0.4
- the CDF is 0.5+1.0+0.4 = 1.9


```python
# getting data of the histogram 
count, bins_count = np.histogram(data_male['Weight'], bins=10) 
print(count) # Frequency of element
print(bins_count) # Each bin values
# finding the PDF of the histogram using count values 
pdf = count / sum(count)  # Pdf used each bin come under at what percentiles
print(pdf)
# using numpy np.cumsum to calculate the CDF 
# We can also find using the PDF values by looping and adding 
cdf = np.cumsum(pdf)  # Cumulative
print(cdf)
# plotting PDF and CDF 
plt.plot(bins_count[1:], pdf, color="red", label="PDF") 
plt.plot(bins_count[1:], cdf, label="CDF") 
plt.legend() 
```

```python
# Change the bin value & check 
count, bins_count = np.histogram(data_male['Weight'], bins=20) 
print(count) # Frequency of element
print(bins_count) # Each bin values
# finding the PDF of the histogram using count values 
pdf = count / sum(count)  # Pdf used each bin come under at what percentiles
print(pdf)
# using numpy np.cumsum to calculate the CDF 
# We can also find using the PDF values by looping and adding 
cdf = np.cumsum(pdf)  # Cumulative
print(cdf)
# plotting PDF and CDF 
plt.plot(bins_count[1:], pdf, color="red", label="PDF") 
plt.plot(bins_count[1:], cdf, label="CDF") 
plt.legend() 
```
