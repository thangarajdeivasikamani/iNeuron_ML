## merging by a different  key

```python
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['A0', 'A1', 'A2', 'A3'],
                     })
   
df2 = pd.DataFrame({'key2': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# what ever left pandas, consider as lef
pd.merge(df1,df2,how='inner',left_on = 'key1',right_on = 'key2')
```

```python
# what ever left pandas, consider as lef
pd.merge(df1,df2,how='left',left_on = 'key1',right_on = 'key2')
```

```python
# Merging same column name
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'B': ['A0', 'A1', 'A2', 'A3'],
                     })
   
df2 = pd.DataFrame({'key2': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# what ever left pandas, consider as lef
pd.merge(df1,df2,how='inner',left_on = 'key1',right_on = 'key2')
```

## merging by a common key

```python
# Based on key
left = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['A0', 'A1', 'A2', 'A3'],
                     })
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# matching key
pd.merge(left,right,how='inner',on = 'key')
```

```python
# All
pd.merge(left,right,how='outer',on = 'key')
```

```python
print("\nAfter simple merging with 'inner' method\n",'-'*50, sep='')

pd.merge(left,right,how='left',on='key')
```

```python
print("\nAfter simple merging with 'inner' method\n",'-'*50, sep='')

pd.merge(left,right,how='right',on='key')
```

```python
# Two keys
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# Default is inner join ( matching keys)- Give as list
pd.merge(left, right, on=['key1', 'key2'])
```

```python
# Left
pd.merge(left, right, how='left',on=['key1', 'key2'])
```

```python
# Right
pd.merge(left, right, how='right',on=['key1', 'key2'])
```

```python
left
```

```python
right
```

```python
# left give more priority
left.join(right)
```

```python
# full join
left.join(right, how='outer')
```

```python
# Define a function
# if more than 500 apply log 10
# or divid by 10
def testfunc(x):
    if (x> 500):
        return (10*np.log10(x))
    else:
        return (x/10)
```
