```python
users['UserID']= users['UserID'].astype(float)
users
```

```python
#12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)
transactions.merge(users,on ='UserID',how = 'left')
```

```python
#13. Which transactions have a UserID not in users?
# https://stackoverflow.com/questions/48647534/python-pandas-find-difference-between-two-data-frames
transactions.merge(users,indicator = True, how='left').loc[lambda x : x['_merge']!='both']
```

```python
#14. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)
transactions.merge(users, on ='UserID', how='inner')
```

```python
# 15. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
transactions.merge(users, on ='UserID', how='outer')
```

```python
#16. Determine which sessions occurred on the same day each user registered
newdf = sessions.merge(users, left_on='SessionDate',right_on='Registered')
newdf
```

```python
#17.Build a dataset with every possible (UserID, ProductID) pair (cross join)
```

```python
# Reference https://www.geeksforgeeks.org/python-program-to-perform-cross-join-in-pandas/
# Now to perform cross join, we will create 
# a key column in both the DataFrames to  
# merge on that key. 
users['key'] = 1
products['key'] = 1
  
# to obtain the cross join we will merge  
# on the key and drop it. 
result = pd.merge(users, products,on ='key').drop("key", axis =1)[['UserID','ProductID',]]
result
```

```python
#18. Determine how much quantity of each product was purchased by each user
pd.merge(result,transactions, how= 'outer').fillna(0)[['UserID','ProductID','Quantity']]
```

```python
#19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)
transactions.merge(transactions,on='UserID')
```

```python
#20. Join each user to his/her first occuring transaction in the transactions table
data = pd.merge(users,transactions,how='left').drop_duplicates(subset ="UserID").reset_index().drop(['key','index'],axis =1)
data 
```

```python
# Find the missing value column
missing_info = list(data.columns[data.isnull().any()])
missing_info
```

```python
# Find the number of missing value rows
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
# Find the missing value %
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
