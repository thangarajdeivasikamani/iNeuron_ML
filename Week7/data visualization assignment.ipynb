```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
X = iris.data
df = pd.DataFrame.from_records(data=iris.data, columns=iris.feature_names)
```

```python
# Convert 4 dimention into 3 dimention using sklearn decomposition.
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)
pca_iris_data = pd.DataFrame(X,columns=['PCA1','PCA2','PCA3'])
pca_iris_data.head()
```

```python
# Method 2 using plotly
import plotly.express as px
fig = px.scatter_3d(pca_iris_data, x=pca_iris_data['PCA1'], y=pca_iris_data['PCA2'], z=pca_iris_data['PCA3'],color=pca_iris_data.columns)
fig.show()
```
