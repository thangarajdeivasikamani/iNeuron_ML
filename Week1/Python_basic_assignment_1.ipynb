```
# Using for loop
result = []
for i in range (2000,3201):
  if (i%7 == 0):
     if (i%5!= 0):
         result.append(i)
for i in result:
  print(i,end=',' )
```

```
# Using list comprehension
result = [i for i in range(2000,3201) if(i%7==0  and i%5 != 0)]
for i in result:
  print(i,end=',' )
```

```
first_name = input('Enter the first Name: ')
second_name = input('Enter the second Name: ')
print('The reverse order of given Name:',first_name[::-1]+' '+second_name[::-1])
```

```
# Using list & revers method
first_name = input('Enter the first Name: ')
second_name = input('Enter the second Name: ')
first_name_list = list(first_name)
first_name_list.reverse()
second_name_list = list(second_name)
second_name_list.reverse()
print('The reverse order of given Name:',''.join(first_name_list)+' '+''.join(second_name_list))
```
