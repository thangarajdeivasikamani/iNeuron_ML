```python
# Add the record
import mysql.connector as connection
# if we are not given database name - It will take first charater as database name
try:
    mydb = connection.connect(host="localhost", database = 'Student',user="root", passwd=passwd_input,use_pure=True)
    # check if the connection is established
    print(mydb.is_connected())
    # Write the table inside the database
    # Column name & Type & length
    # We need to define schema before push
    # Column name : Studentid,FirstName,LastName,RegistrationDate,Class,Section
    # AUTO_INCREMENT PRIMARY KEY  - Create the auto increament id - Primary key means - unique id
    # VARCHAR - Equal to string type
    query = "CREATE TABLE StudentDetails (Studentid INT(10) AUTO_INCREMENT PRIMARY KEY,FirstName VARCHAR(60)," \
            "LastName VARCHAR(60), RegistrationDate DATE,Class Varchar(20), Section Varchar(10))"

    cursor = mydb.cursor() #create a cursor to execute queries
    # Push or pull the data
    cursor.execute(query)
    print("Table Created!!")
    mydb.close()
except Exception as e:
    mydb.close()
    print(str(e))
```

```python
# Insert from file 
import mysql.connector as connection
import pandas as pandas
import csv

try:
    mydb = connection.connect(host="localhost", user="root", passwd=passwd_input,use_pure=True)
    #check if the connection is established
    print(mydb.is_connected())
    #create a new database
    query = "Create database GlassData;"
    cursor = mydb.cursor() #create a cursor to execute queries
    cursor.execute(query)
    print("Database Created!!")
    mydb.close() #close the connection

    #Establish a new connection to the database created above
    mydb = connection.connect(host="localhost", database = 'GlassData',user="root", passwd=passwd_input, use_pure=True)

    #create a new table to store glass data if not exist
    # Here float 10 whole number , 5 decimal number
    query = "CREATE TABLE IF NOT EXISTS GlassData (Index_Number INT(10),RI float(10,5), Na float(10,5), Mg float(10,5),Al float(10,5)," \
            " Si float(10,5), K float(10,5), Ca float(10,5), Ba float(10,5), Fe float(10,5), Class INT(5))"
    cursor = mydb.cursor()  # create a cursor to execute queries
    cursor.execute(query)
    print("Table Created!!")

    #read from the file
    with open('glass.data', "r") as f:
        # Navigate the pointer to nextt line ( Igonre the first line)
        next(f)
        # read using csv reader
        glass_data = csv.reader(f, delimiter="\n")
        # Enumerate key & value 
        for line in enumerate(glass_data):
            # Push only values, list_ will take multiple tuple values
            for list_ in (line[1]):
                cursor.execute('INSERT INTO GlassData values ({values})'.format(values=(list_)))
    print("Values inserted!!")
    mydb.commit()
    cursor.close()
    mydb.close()

except Exception as e:
    #mydb.close()
    print(str(e))
```

```python
# Group by
import mysql.connector as connection
import pandas as pandas

try:

    mydb = connection.connect(host="localhost", database='GlassData', user="root", passwd=passwd_input, use_pure=True)
    # check if the connection is established
    print(mydb.is_connected())
    query = "select si, Count(Si),Class from GlassData.GlassData group by Class order by Si;" #group by class
    groupby_result_dataFrame = pandas.read_sql(query,mydb)
    #print(groupby_result_dataFrame)

    mydb.close()  # close the connection

except Exception as e:
    #mydb.close()
    print(str(e))
```
